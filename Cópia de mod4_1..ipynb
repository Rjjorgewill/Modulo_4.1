```python
import pandas as pd
import numpy as np


tabela = pd.DataFrame({'Dia': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'],
                      'Limpeza': [100, 0, 100, 0, 100, 100, 0],
                      'Comida': [221.60, 375.31, 412, 495.20, 441.53, 245, 164],
                      'Transporte': [150, 100, 125, 300, 275, 525, 75],
                      'Outros': [0, 0, 2310, 500, 0, 0, 820],
                      'Receita': [2200, 2420.50, 3391, 5322, 4898.50, 4200, 3893]})
tabela
```

```python
tabela['Receita'] = tabela['Receita'] * 0.93

tabela['Receita']
```

```python
soma_receita = tabela['Receita'].sum()

print("Soma da receita semanal: %.2f"% soma_receita)
```

```python
media_receita = tabela['Receita'].mean()

print("Media da receita semanal: %.2f"% media_receita)
```

```python
print('Soma dos custos da limpeza:',tabela['Limpeza'].sum())

print('Soma dos custos da comida:',tabela['Comida'].sum())

print('Soma dos custos do transporte:',tabela['Transporte'].sum())

print('Soma dos custos de outros:',tabela['Outros'].sum())
```

```python
mediasem = tabela['Limpeza'].sum() + tabela['Comida'].sum() + tabela['Transporte'].sum() + tabela['Outros'].sum() 

print('A media de despesas foi de: {:.2f}'.format(mediasem/4))
```

```python
tabela['Lucro'] = tabela['Receita'] - (tabela['Limpeza'] + tabela['Comida'] + tabela['Transporte'] + tabela['Outros'])

melhor_Dia = tabela['Dia'][tabela['Lucro'].idxmax()]

soma_Lucro = tabela['Lucro'].sum()

print(f"Dia mais lucrativo: {melhor_Dia} e seu lucro foi de:{tabela['Lucro'].max():.2f}")
print(f"Lucro total da semana foi:{soma_Lucro:.2f}")

tabela

```
